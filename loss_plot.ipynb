```python
# this batch is for plotting the loss curve
# /home_data/home/linxin2024/code/3DMedDM_v2/save/_train_lccd_sr/log.txt

import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["font.sans-serif"]=["SimHei"] #设置字体
plt.rcParams["axes.unicode_minus"]=False #正常显示负号

# 设置文件路径
log_file_path = '/home_data/home/linxin2024/code/3DMedDM_v2/save/_train_lccd_sr/log.txt'

# 初始化存储数据的列表
epochs = []
loss0 = []
loss1 = []

# 读取日志文件
with open(log_file_path, 'r') as file:
    for line in file:
        # 检查是否是包含loss信息的行
        if 'epoch' in line and 'loss0' in line and 'loss1' in line:
            parts = line.split(',')
            epoch_part = parts[0].split('/')[0].strip().split(' ')[1]  # 提取epoch
            loss0_part = float(parts[2].split('=')[1].strip())  # 提取loss0
            loss1_part = float(parts[3].split('=')[1].strip())  # 提取loss1
            
            epochs.append(int(epoch_part))
            loss0.append(loss0_part)
            loss1.append(loss1_part)

# 转换为numpy数组
epochs = np.array(epochs)
loss0 = np.array(loss0)
loss1 = np.array(loss1)
```

```python
color1 = (230,133,127)   # 红
color2 = (60,123,192)     # 蓝

# 将RGB颜色转换为0到1之间的浮点数  
color1_normalized = tuple(c / 255.0 for c in color1)  
color2_normalized = tuple(c / 255.0 for c in color2) 
```

```python
# 绘制曲线图
plt.figure(figsize=(10, 6))
plt.plot(epochs, loss0, label='Loss0', color=color1_normalized, marker='o')
plt.title('Loss Curve')
plt.xlabel('Epoch')
plt.ylabel('Loss0')
# plt.legend()
plt.grid()
plt.xticks(np.arange(0, max(epochs)+1, 50))  # 设置x轴刻度
plt.tight_layout()
plt.show()  # 显示图表
```

```python
# 绘制曲线图
plt.figure(figsize=(10, 6))
plt.plot(epochs, loss1, label='Loss1', color=color2_normalized, marker='o')
plt.title('Loss Curve')
plt.xlabel('Epoch')
plt.ylabel('Loss1')
# plt.legend()
plt.grid()
plt.xticks(np.arange(0, max(epochs)+1, 50))  # 设置x轴刻度
plt.tight_layout()
plt.show()  # 显示图表
```
